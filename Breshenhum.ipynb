```python
import matplotlib.pyplot as plt
```

```python
def line_draw_bresenhum(x1, y1, x2, y2):
  dx=x2-x1
  dy=y2-y1

  p=2*dy-dx
  x,y=x1,y1
  line_x=[x]
  line_y=[y]
  while(x < x2):
    if p<0:
      x+=1
     
      line_x.append(x)
      line_y.append(y)
      p=p+2*dy
      
    else:
      x+=1
      y+=1
      
    line_x.append(x)
    line_y.append(y)
    p=p+2*dy-2*dx

    return line_x,line_y
  
  #use example
  x1,y1=9,18
  x2,y2=16,22
  line_x, line_y=line_draw_bresenhum(x1,y1,x2,y2)

plt.plot(line_x, line_y,marker='o',color='r')
plt.title('Line Drawing using DDA')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()




```
