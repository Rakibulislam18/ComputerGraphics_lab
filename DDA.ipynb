```python
import matplotlib.pyplot as plt
```

```python
def Line_drawing_DDA(x1, y1, x2, y2):
    dx=x2 - x1
    dy=y2 - y1

    steps=max(abs(dx),abs(dy))
    m=dy/dx
    x=x1
    y=y1
    line_x=[x]
    line_y=[y]

    if(abs(dx)>abs(dy)):
        for i in range(steps):
            x=x+1
            y=y+m
            line_x.append(x)
            line_y.append(y)
    else:
        for i in range(steps):
            x=x+1/m
            y=y+1
            line_x.append(x)
            line_y.append(y)
    return line_x, line_y

x1, y1 = 1, 1
x2, y2 = 8, 4
line_x, line_y=Line_drawing_DDA(x1, y1, x2, y2)
plt.plot(line_x, line_y,marker='o',color='r')
plt.title('Line Drawing using DDA')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()


   
```
